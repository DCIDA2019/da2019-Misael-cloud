# como filtrar renglones por valor de columna

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies.shape
```

```python
booleans = []
for i in movies.duration:
    if i >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
booleans[0:5]
```

```python
len(booleans)
```

# convert booleans list to a Pandas Series

```python
type(booleans)
```

```python
is_long = pd.Series(booleans)
```

```python
is_long.head()
```

```python
movies[is_long]
#solo te muestra los True
```

```python
is_long = movies.duration >= 200
is_long.head()
```

```python
movies[movies.duration >= 200]
```

```python
movies[movies.duration >= 200]['genre']
```

```python
movies.loc[movies.duration >= 200, 'genre']
#mucho mejor que la anterior
```

```python
movies['D'] = is_long
movies.head()
```

# Múltiples Filtros en un DataFrame

```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
```

```python
movies.genre.isin(['Crime', 'Drama', 'Action'])
#hace la misma funcion que un | una conjuncion, o sea que aparezco Crime o Drama o Action
```

```python
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])]
```

# Cuando usar groupby

```python
drinks = pd.read_csv('http://bit.ly/drinksbycountry')
drinks.head()
```

```python
drinks.beer_servings.mean()
```

# promedio por region de continente

```python
drinks.groupby('continent').beer_servings.mean()
```

```python
drinks[drinks.continent == 'Africa'].beer_servings.mean()
#otra forma en vez de groupby
```

```python
drinks.groupby('continent').beer_servings.max()
```

```python
drinks.groupby('continent').beer_servings.agg(['count','min','max','mean'])
```

```python
drinks.groupby('continent').mean()
```

```python
%matplotlib inline
```

```python
drinks.groupby('continent').mean().plot(kind='bar')
```
