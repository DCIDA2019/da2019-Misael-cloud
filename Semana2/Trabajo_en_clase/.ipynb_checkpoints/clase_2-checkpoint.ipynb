```python
def f_1(x):
    a = []
    for i in range(len(x)):
        if x[i]%2 == 0:
            a.append(x[i]**2)
        else:
            a.append(np.log(x[i]))
    return np.array(a)
```

```python
np_array_1 = np.arange(1,50,3)
np_array_1
```

```python
f_1(np_array_1)
```

```python
def f(x):
    if x%2==0:
        return x**2
    else:
        return np.log(x)
    
f_vectorizado = np.vectorize(f)
f_vectorizado(np_array_1)
```

```python
x_1 = np.linspace(-2*(np.pi), 2*(np.pi), 50)
x_1
```

```python
def g(x):
    if x>0:
        return np.cos(x)
    else:
        return np.sin(x)

g1=np.vectorize(g)
y2 = g1(x_1)
```

```python
class funciones():
    
    def __init__(self,A,saludo='Hola'):
        self.A =A
        self.saludo=saludo
        
    def g(self,x):

        return self.A*np.cos(x),self.A*np.sin(x)
        

    def f2(self,x):
        if x%2==0:
            return x*x
        else:
            return np.log(x)
        
    def f(self):
        return self.saludo
```

```python
x=funciones(3,'Hello')
```

Descargar datos de una página web

```python
import requests
import pandas as pd
import os
```

```python
url='https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007'
html=requests.get(url).content
```

```python
df_list=pd.read_html(html,header=1)
df=df_list[-1]
df.to_csv('salarios_minimos.csv')
```

```python
file='salarios_minimos.csv'
if not os.path.isfile(file):
    url='https://idconline.mx/laboral/salarios-minimos/salarios-minimos-1986-2007'
    html=requests.get(url).content
    df_list=pd.read_html(html,header=1)
    df=df_list[-1]
    df.to_csv('salarios_minimos.csv')
else:
    df = pd.read_csv(file)
    
df
```
